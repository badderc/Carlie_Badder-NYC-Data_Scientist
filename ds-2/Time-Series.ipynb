### Check Stationarity
ie. is the mean and variance constant over time

```python
# Take the moving window average over a year of data (daily)
rolmean = pd.rolling_mean(df_recent.value, window=365) # for days in a year
rolstd = pd.rolling_std(df_recent.value, window=365)
```

```python
#Plot rolling statistics:
import matplotlib.pyplot as plt

x_a = range(len(df_recent.value))
plt.figure(figsize=(10,8))
orig = plt.plot(x_a, df_recent.value, color='blue',label='Original')
mean = plt.plot(x_a, rolmean, color='red', label='Rolling Mean')
std = plt.plot(x_a, rolstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')

# The mean is increasing with time, so this is not a stationary series
# Need to reduce trend and/or seasonality
```

### Trend
Remove trend

```python
# Take the log to remove trend
ws = df_recent.value
ws_log = np.log(ws)
```

```python
plt.figure(figsize=(10,8))

plt.plot(x_a, ws_log)
plt.title('Log of Time Series to Remove Trend')
plt.xlabel('Lag')
plt.ylabel('Wilshire REIT Index')

# Still has trend
```

```python
# Try moving window average (mwa)
ws_log_mwa = pd.rolling_mean(ws_log, 365) # to average over a year

plt.figure(figsize=(10,8))
plt.plot(x_a, ws_log)
plt.plot(x_a, ws_log_mwa, color='red')
plt.title('Log of Time Series with Moving Window Average to Remove Trend')
plt.xlabel('Lag')
plt.ylabel('Wilshire REIT Index')
plt.legend(['WS', 'Log WS MWA'])
```

```python
# Moving window average difference
ws_log_mwa_diff = ws_log - ws_log_mwa
ws_log_mwa_diff.dropna(inplace=True)

rolmean2 = pd.rolling_mean(ws_log_mwa_diff, window=365) # for 52 weeks in a year
rolstd2 = pd.rolling_std(ws_log_mwa_diff, window=365)

x_d = range(len(ws_log_mwa_diff))
plt.figure(figsize=(10,8));
orig2 = plt.plot(x_d, ws_log_mwa_diff, color='blue',label='Original')
mean2 = plt.plot(x_d, rolmean2, color='red', label='Rolling Mean')
std2 = plt.plot(x_d, rolstd2, color='black', label = 'Rolling Std')
plt.legend(loc='best');
plt.title('Rolling Mean & Standard Deviation');
```

```python
# Mean and standard deviation are more constant now
# ws_log_mwa_diff
```

### Seasonality
Could use several other methods to reduce seasonality beyond taking the log difference, but this could be explored in later iterations beyond the MVP

```python
# Reduce trend and seasonality by taking the log and the shift difference 
ws_log_diff = ws_log - ws_log.shift()

plt.figure(figsize=(10,8))
plt.plot(x_a, ws_log_diff)
plt.title('Wilshire REIT Index with Seasonality and Trend Reduction')
```

```python
# Check stationarity

ws_log_diff.dropna(inplace=True)

rolmean3 = pd.rolling_mean(ws_log_diff, window=365) # for days in a year
rolstd3 = pd.rolling_std(ws_log_diff, window=365)

x_ld = range(len(ws_log_diff))
plt.figure(figsize=(10,8));
orig2 = plt.plot(x_ld, ws_log_diff, color='blue',label='Original')
mean2 = plt.plot(x_ld, rolmean3, color='red', label='Rolling Mean')
std2 = plt.plot(x_ld, rolstd3, color='black', label = 'Rolling Std')
plt.legend(loc='best');
plt.title('Rolling Mean & Standard Deviation of Log Diff of WS REIT Index');
```

### Autoregression with WS Log Diff

* Using ARIMA (Autoregressive Integrated Moving Averages) - similar to a linear regression but with dependent observations
* AR Terms - number of lags to use to predict the next value
* MA Terms - number of values to use in moving averages
* d - number of difference - ws_log_diff uses a first order difference, but could take more than one

```python
#ACF (autocorrelation function) and PACF (partial autocorrelation function) plots:
# Use plots to get AR and MA values
from statsmodels.tsa.stattools import acf, pacf
```

```python
lag_acf = acf(ws_log_diff, nlags=20)
lag_pacf = pacf(ws_log_diff, nlags=20, method='ols')
```

```python
plt.figure(figsize=(15,6))
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ws_log_diff)),linestyle='--',color='gray') # confidence interval
plt.axhline(y=1.96/np.sqrt(len(ws_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
plt.xticks(np.arange(0, 20, step=1));

#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ws_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ws_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
plt.xticks(np.arange(0, 20, step=1));
```

```python
# From where the plots cross the upper bounds of the confidence intervals
# q (number of AR terms) = 1 and p (number of MA terms) = 1
```

```python
from statsmodels.tsa.arima_model import ARIMA

# Just Autoregression Model
model = ARIMA(ws_log, order=(1, 1, 0))  
results_AR = model.fit(disp=-1)  

plt.figure(figsize=(10,8))
plt.plot(x_ld, ws_log_diff)
plt.plot(x_ld, results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-ws_log_diff)**2))
plt.legend(['WS Log Diff', 'AR Results']);
```

```python
# Just Moving Averages Model

model = ARIMA(ws_log, order=(0, 1, 1))  
results_MA = model.fit(disp=-1)

plt.figure(figsize=(10,8))
plt.plot(x_ld, ws_log_diff)
plt.plot(x_ld, results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-ws_log_diff)**2))
plt.legend(['WS Log Diff', 'MA Results']);
```

```python
# Full ARIMA Model

model = ARIMA(ws_log, order=(1, 1, 1))  
results_ARIMA = model.fit(disp=-1)

plt.figure(figsize=(10,8))
plt.plot(x_ld, ws_log_diff)
plt.plot(x_ld, results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ws_log_diff)**2))
plt.legend(['WS Log Diff', 'ARIMA Results']);
```
